# Train model

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, SimpleRNN, GRU
import numpy as np
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
scaled_data1_price_change = scaler.fit_transform((data1['close'] - data1['open']).values.reshape(-1,1))
scaled_data2_price_change = scaler.fit_transform((data2['close'] - data2['open']).values.reshape(-1,1))
scaled_data3_price_change = scaler.fit_transform((data3['close'] - data3['open']).values.reshape(-1,1))
```

# Stock 1

```python
# need change
scaled_data = scaled_data1_price_change


prediction_days = 10

x_train = []
y_train = []

for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x-prediction_days:x,0])
    y_train.append(scaled_data[x,0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
model = Sequential()
model.add(GRU(3, return_sequences=True, input_shape=(x_train.shape[1], 1) ))
model.add(GRU(3, return_sequences=False))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs = 50, batch_size =32)
```

```python
model.save("models/Price_of_change/RNN/EURUSD.h5")
```
